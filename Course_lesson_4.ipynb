<h3 align='center'>Вивести матрицю 1</h3>
На вхід до програми подаються два натуральні числа n і m, кожне на окремому рядку — кількість рядків і стовпців у матриці. Далі вводяться самі елементи матриці - слова, кожне на окремому рядку; поспіль йдуть елементи спочатку першого рядка, потім другого, і т.д.

Напишіть програму, яка спочатку зчитує елементи матриці один за одним, а потім виводить їх у вигляді матриці.

<b>Формат вхідних даних</b>
На вхід до програми подаються два числа n і m — кількість рядків і стовпців у матриці, далі йдуть n×m слів, кожне на окремому рядку.

<b>Формат вихідних даних</b>
Програма повинна вивести лічену матрицю, розділяючи її елементи одним пробілом.

```python
# put your python code here
n = int(input())
m = int(input())

matrix = [[0]*m for _ in range(n)] 
for i in range(n):
    for j in range(m):
        text = input()
        matrix[i][j] = text

print('')
for i in range(n):
    print(*matrix[i])
   
print('')
for i in range(m):
    for j in range(n):
        print(*matrix[j][i])
```

<h3 align='center'>Слід матриці</h3>
Після квадратної матриці називається сума елементів головної діагоналі. Напишіть програму, яка виводить слід заданої квадратної матриці.

<b>Формат вхідних даних</b><br>
На вхід до програми подається натуральне число nn — кількість рядків і стовпців у матриці, потім елементи матриці (цілі числа) рядково через пропуск.

<b>Формат вихідних даних</b><br>
Програма має вивести одне число – слід заданої матриці.

```python
# put your python code here
n = int(input())
result = 0

matrix = [[0]*n for _ in range(n)] 

for i in range(n):
    text = input().split(' ')
    for j in range(n):
        matrix[i][j] = text[j]

for i in range(n):
    if i > 0:
        break
    for j in range(n):
        result += int(matrix[j][j])
print('')
print(result)
```

<h3 align='center'>Більше середнього</h3>
Напишіть програму, яка виводить кількість елементів квадратної матриці в кожному рядку, більших за середній арифметичний елементів даного рядка.

<b>Формат вхідних даних</b><br>
На вхід до програми подається натуральне число n — кількість рядків і стовпців у матриці, потім елементи матриці (цілі числа) рядково через пропуск.

<b>Формат вихідних даних</b><br>
Програма повинна вивести n чисел — для кожного рядка кількість елементів матриці, більших за середній арифметичний елементів даного рядка.

```python
# put your python code here
import numpy as np
matrix = [list(map(int, input().split())) for i in range(int(input()))]
print('')
for i in range(len(matrix)):
    result = 0
    for j in range(len(matrix)):
        if matrix[i][j] > np.mean(matrix[i]):
            result+=1
        if j == len(matrix)-1:
            print(result)
```

<h3 align='center'>Максимальний в області 1</h3>
Напишіть програму, яка виводить максимальний елемент у заштрихованій області квадратної матриці.




<b>Формат вхідних даних</b><br>
На вхід до програми подається натуральне число nn — кількість рядків і стовпців у матриці, потім елементи матриці (цілі числа) рядково через пропуск.

<b>Формат вихідних даних</b><br>
Програма повинна вивести одне число - максимальний елемент у області квадратної матриці.

<b>Примітка</b>. Елементи головної діагоналі також враховуються!

```python
matrix = [list(map(int, input().split())) for i in range(int(input()))]
lst = []
j = 0
maxi = -999
for i in range(len(matrix)):
    lst.append(matrix[i][:j+1])
    j+=1
for i in range(len(lst)):
    for j in range(len(lst[i])):
        if lst[i][j] > maxi:
            maxi = lst[i][j]
print('')
print(maxi)
```

<b>Формат вхідних даних</b><br>
На вхід до програми подається натуральне число nn — кількість рядків і стовпців у матриці, потім елементи матриці (цілі числа) рядково через пропуск.

<b>Формат вихідних даних</b><br>
Програма повинна вивести одне число - максимальний елемент у області квадратної матриці.

<b>Примітка</b>. Елементи діагоналей також враховуються.

```python
# put your python code here
matrix = [list(map(int, input().split())) for i in range(int(input()))]
lst = []
for i in range(len(matrix)):
    for j in range(len(matrix)):
        if ((i>=j) and (i<=len(matrix)-1-j)) or ((i<=j) and (i>=len(matrix)-1-j)):
            lst.append(matrix[i][j])
print('')
print(max(lst))
```

<h3 align='center'>Суми чвертей</h3>
Квадратна матриця розбивається на чотири чверті, обмежені головною та побічною діагоналями: верхню, нижню, ліву та праву.



Напишіть програму, яка обчислює суму елементів верхньої чверті; правої чверті; нижньої чверті; лівої чверті.

<b>Формат вхідних даних</b><br>
На вхід до програми подається натуральне число nn — кількість рядків і стовпців у матриці, потім елементи матриці (цілі числа) рядково через пропуск.

<b>Формат вихідних даних</b><br>
Програма має вивести текст відповідно до умови завдання.

<b>Примітка</b>. Елементи діагоналей не враховуються.

```python
# put your python code here
matrix = [list(map(int, input().split())) for i in range(int(input()))]
up = right = left = down = 0
for i in range(len(matrix)):
    for j in range(len(matrix)):
        if ((i<j) and (i<len(matrix)-1-j)): # up
            up += matrix[i][j]
        elif ((i<j) and (i>len(matrix)-1-j)): # right
            right += matrix[i][j]
        elif ((i>j) and (i>len(matrix)-1-j)): # down
            down += matrix[i][j]
        elif ((i>j) and (i<len(matrix)-1-j)): # left
            left += matrix[i][j]
print('')
print(f'Верхняя четверть: {up}')
print(f'Правая четверть: {right}')
print(f'Нижняя четверть: {down}')
print(f'Левая четверть: {left}')
```
