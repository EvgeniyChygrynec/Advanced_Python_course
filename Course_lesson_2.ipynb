<h3 align='center'>Координатні чверті</h3>

Даний набір точок на координатній площині. Необхідно підрахувати та вивести кількість точок, що лежать у кожній координатній чверті.

<b>Формат вхідних даних</b><br>
У першому рядку записано кількість точок. Кожен наступний рядок складається з двох цілих чисел - координат точки (спочатку абсцис - x, потім ордината - y), розділених символом пробілу.

<b>Формат вихідних даних</b><br>
Програма повинна вивести кількість точок, що лежать у кожній координатній чверті, як у прикладах.

<b>Примітка</b>. Зверніть увагу, що точки, що лежать на осях координат, не прийнято відносити до будь-якої координатної чверті.

```python
# put your python code here
n = int(input())
n1 = n2 = n3 = n4 = 0
for _ in range(n):
    cord = input()
    cord = cord.split(' ')
    x = int(cord[0])
    y = int(cord[1])
    if x > 0 and y > 0:
        n1 += 1
    elif x < 0 and y > 0:
        n2 += 1
    elif x < 0 and y < 0:
        n3 += 1
    elif x > 0 and y < 0:
        n4 += 1
        
print('')
print(f'Первая четверть: {n1}')
print(f'Вторая четверть: {n2}')
print(f'Третья четверть: {n3}')
print(f'Четвертая четверть: {n4}')
```

<h3 align='center'>Більше попереднього</h3>
На вхід до програми подається рядок тексту з натуральних чисел. З неї формується перелік чисел. Напишіть програму підрахунку кількості чисел, які більше попереднього їм у цьому списку числа, тобто стоять за меншим числом.

<b>Формат вхідних даних</b><br>
На вхід до програми подається рядок тексту з розділених пробілами натуральних чисел.

<b>Формат вихідних даних</b><br>
Програма має вивести одне число – кількість елементів списку, більших за попередній.

```python
# put your python code here
n = input().split(' ')
counter = 0
for i in range(len(n)):
    if i == 0:
        continue
    elif int(n[i]) > int(n[i-1]):
        counter += 1
    else:
        continue
print('')
print(counter)
```

<h3 align='center'>Назад, вперед і навпаки</h3>
На вхід до програми подається рядок тексту з натуральних чисел. З елементів рядка формується перелік чисел. Напишіть програму, яка змінює місцями сусідні елементи списку (a[0] c a[1], a[2] c a[3] тощо). Якщо у списку непарна кількість елементів, то останній залишається на своєму місці.

<b>Формат вхідних даних</b><br>
На вхід до програми подається рядок тексту, що містить натуральні числа, розділені пробілами.

<b>Формат вихідних даних</b><br>
Програма повинна вивести змінений список, розділяючи його елементи одним пробілом.

```python
# put your python code here
n = input().split(' ')

for i in range(len(n)):
    if i%2 == 0:
        continue
    elif i%2 != 0:
        n[i], n[i-1] = n[i-1], n[i]

print('')
print(*n)
```

<h3 align='center'>Добуток чисел</h3>
Напишіть програму для визначення, чи є числом твором двох чисел з цього набору, що виводить результат у вигляді відповіді «ТАК» або «НІ».

<b>Формат вхідних даних</b><br>
У першому рядку подається число n, (0 < n < 1000) n (0 < n < 1000) – кількість чисел у наборі. У наступних nn рядках вводяться цілі числа, що становлять набір (можуть повторюватися). Потім слідує ціле число, яке є чи не є добутком двох якихось чисел із набору.

<b>Формат вихідних даних</b><br>
Програма повинна вивести «ТАК» або «НІ» відповідно до умови завдання.

<b>Примітка 1</b>. Саме на себе число з набору помножитися не може, тобто два множники повинні мати різні індекси в наборі.

<b>Примітка 2</b>. Використовуйте вкладені цикли для вирішення завдання.

```python
# put your python code here
n = input()
lista = []
lista2 = []
for _ in range(int(n)):
    m = int(input())
    lista.append(m)

for i in range(len(lista)):
    for j in range(len(lista)): 
        if i != j:
            lista2.append(lista[i]*lista[j])

k = input()
if int(k) in lista2:
    print('ДА')
else:
    print('НЕТ')
```

<h3 align='center'>Камень, ножницы, бумага</h3>
Тимур и Руслан пытаются разделить фронт работы по курсу "Python для профессионалов". Для этого они решили сыграть в камень, ножницы и бумагу. Помогите ребятам бросить честный жребий и определить, кто будет делать очередной модуль нового курса.

<b>Формат входных данных</b><br>
На вход программе подаются две строки текста, содержащие слова "камень", "ножницы" или "бумага". На первой строке записан выбор Тимура, на второй – выбор Руслана.

<b>Формат выходных данных</b><br>
Программа должна вывести результат жеребьёвки, то есть кто победит: Тимур, Руслан или же они сыграют вничью.

<b>Примечание</b>. Правила игры стандартные: камень побеждает ножницы, но проигрывает бумаге, а ножницы побеждают бумагу.

```python
# put your python code here
T = input()
R = input()
print('')
if T == 'камень':
    if R == T:
        print('ничья')
    elif R == 'бумага':
        print('Руслан')
    else:
        print('Тимур')
elif T == 'бумага':
    if R == T:
        print('ничья')
    elif R == 'ножницы':
        print('Руслан')
    else:
        print('Тимур')
elif T == 'ножницы':
    if R == T:
        print('ничья')
    elif R == 'камень':
        print('Руслан')
    else:
        print('Тимур')
```

<b>Формат вихідних даних</b><br>
Програма має вивести результат жеребкування: хто переміг – Тимур чи Руслан, чи вони зіграли внічию.

<b>Примітка</b>. Правила гри стандартні: ножиці ріжуть папір. Папір загортає камінь. Камінь тисне ящірку, а ящірка труїть Спока, у той час як Спок ламає ножиці, які, у свою чергу, відрізають голову ящірці, яка їсть папір, на якому докази проти Спока. Спок випаровує камінь, а камінь, зрозуміло, затупляє ножиці.

```python
# put your python code here
T = input()
R = input()
print('')
game = {
'камень': ['ящерица', 'ножницы'],
'ножницы': ['ящерица', 'бумага'],
'бумага': ['Спок', 'камень'],
'Спок': ['камень', 'ножницы'],
'ящерица': ['Спок', 'бумага']
}
if R == T:
    print('ничья')
elif R in game[T]:
    print('Тимур')
else:
    print('Руслан')
```
