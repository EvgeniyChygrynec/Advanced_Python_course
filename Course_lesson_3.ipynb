–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ø–æ–¥–∞—î—Ç—å—Å—è —á–∏—Å–ª–æ nn. –ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ –ø–æ–≤–µ—Ä—Ç–∞—î –≤–∫–∞–∑–∞–Ω–∏–π —Ä—è–¥–æ–∫ —Ç—Ä–∏–∫—É—Ç–Ω–∏–∫–∞ –ü–∞—Å–∫–∞–ª—è —É –≤–∏–≥–ª—è–¥—ñ —Å–ø–∏—Å–∫—É (–Ω—É–º–µ—Ä–∞—Ü—ñ—è —Ä—è–¥–∫—ñ–≤ –ø–æ—á–∏–Ω–∞—î—Ç—å—Å—è –∑ –Ω—É–ª—è).

<b>–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ø–æ–¥–∞—î—Ç—å—Å—è —á–∏—Å–ª–æ n (n‚â•0).

<b>–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ü—Ä–æ–≥—Ä–∞–º–∞ –º–∞—î –≤–∏–≤–µ—Å—Ç–∏ –≤–∫–∞–∑–∞–Ω–∏–π —Ä—è–¥–æ–∫ —Ç—Ä–∏–∫—É—Ç–Ω–∏–∫–∞ –ü–∞—Å–∫–∞–ª—è —É –≤–∏–≥–ª—è–¥—ñ —Å–ø–∏—Å–∫—É.

<b>–ü—Ä–∏–º—ñ—Ç–∫–∞ 1</b>. –†—ñ—à–µ–Ω–Ω—è –∑—Ä—É—á–Ω–æ –æ—Ñ–æ—Ä–º–∏—Ç–∏ —É –≤–∏–≥–ª—è–¥—ñ —Ñ—É–Ω–∫—Ü—ñ—ó pascal(), —è–∫–∞ –ø—Ä–∏–π–º–∞—î –Ω–æ–º–µ—Ä —Ä—è–¥–∫–∞ —Ç–∞ –ø–æ–≤–µ—Ä—Ç–∞—î –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏–π —Ä—è–¥–æ–∫ —Ç—Ä–∏–∫—É—Ç–Ω–∏–∫–∞ –ü–∞—Å–∫–∞–ª—è.

<b>–ü—Ä–∏–º—ñ—Ç–∫–∞ 2</b>. –ì—Ä–∞—Ñ—ñ—á–Ω–∞ —ñ–ª—é—Å—Ç—Ä–∞—Ü—ñ—è —Ñ–æ—Ä–º—É–≤–∞–Ω–Ω—è —Ç—Ä–∏–∫—É—Ç–Ω–∏–∫–∞ –ü–∞—Å–∫–∞–ª—è.

```python
from math import factorial

n = (int(input())+1)

my_list = [0] * (n)
    
for i in range(n):
    for j in range(i+1):
        if i == (n-1):
            my_list[j] = (factorial(i)//(factorial(j)*factorial(i-j)))
            
print(my_list)
```

<h3 align='center'>–¢—Ä–∏–∫—É—Ç–Ω–∏–∫ –ü–∞—Å–∫–∞–ª—è 2</h3>
–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ø–æ–¥–∞—î—Ç—å—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–µ —á–∏—Å–ª–æ n. –ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ –≤–∏–≤–æ–¥–∏—Ç—å –ø–µ—Ä—à—ñ n —Ä—è–¥–∫—ñ–≤ —Ç—Ä–∏–∫—É—Ç–Ω–∏–∫–∞ –ü–∞—Å–∫–∞–ª—è.

<b>–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ø–æ–¥–∞—î—Ç—å—Å—è —á–∏—Å–ª–æ n (n‚â•1).

<b>–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ü—Ä–æ–≥—Ä–∞–º–∞ –ø–æ–≤–∏–Ω–Ω–∞ –≤–∏–≤–µ—Å—Ç–∏ –ø–µ—Ä—à—ñ nn —Ä—è–¥–∫—ñ–≤ —Ç—Ä–∏–∫—É—Ç–Ω–∏–∫–∞ –ü–∞—Å–∫–∞–ª—è, –∫–æ–∂–µ–Ω –Ω–∞ –æ–∫—Ä–µ–º–æ–º—É —Ä—è–¥–∫—É –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ –∑—Ä–∞–∑–∫–∞.

```python
from math import factorial 

n = int(input())
for i in range(n):
    for j in range(i+1):
        print(factorial(i)//(factorial(j)*factorial(i-j)), end=" ")
    print()
```

<h3 align='center'>–£–ø–∞–∫–æ–≤–∫–∞ –¥—É–±–ª—ñ–∫–∞—Ç—ñ–≤ üå∂Ô∏è</h3>
–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ø–æ–¥–∞—î—Ç—å—Å—è —Ä—è–¥–æ–∫ —Ç–µ–∫—Å—Ç—É, —â–æ –º—ñ—Å—Ç–∏—Ç—å —Å–∏–º–≤–æ–ª–∏. –ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ –ø–∞–∫—É—î –ø–æ—Å–ª—ñ–¥–æ–≤–Ω–æ—Å—Ç—ñ –æ–¥–Ω–∞–∫–æ–≤–∏—Ö —Å–∏–º–≤–æ–ª—ñ–≤ –∑–∞–¥–∞–Ω–æ–≥–æ —Ä—è–¥–∫–∞ –≤ –ø—ñ–¥—Å–ø–∏—Å–∫–∏.

<b>–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ø–æ–¥–∞—î—Ç—å—Å—è —Ä—è–¥–æ–∫ —Ç–µ–∫—Å—Ç—É, —â–æ –º—ñ—Å—Ç–∏—Ç—å —Å–∏–º–≤–æ–ª–∏, –≤—ñ–¥–æ–∫—Ä–µ–º–ª–µ–Ω—ñ —Å–∏–º–≤–æ–ª–æ–º –ø—Ä–æ–ø—É—Å–∫—É.

<b>–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ü—Ä–æ–≥—Ä–∞–º–∞ –º–∞—î –≤–∏–≤–µ—Å—Ç–∏ –≤–∫–∞–∑–∞–Ω–∏–π —Å–ø–∏—Å–æ–∫.

```python
# put your python code here
text = input().split(' ')
lst = [['']]
for i, item in enumerate(text):
    if lst[-1][0] == item:
        lst[-1].append(item)
    else:
        lst.append([item])
lst.remove([''])
print(lst)
```

<h3 align='center'>–†–æ–∑–±–∏—Ç—Ç—è –Ω–∞ —á–∞–Ω–∫–∏ üå∂Ô∏è</h3>
–ù–∞ –≤—Ö—ñ–¥ –ø—Ä–æ–≥—Ä–∞–º—ñ –ø–æ–¥–∞—é—Ç—å—Å—è –¥–≤–∞ —Ä—è–¥–∫–∏, –æ–¥–Ω—ñ—î—é —Å–∏–º–≤–æ–ª–∏, –≤ —ñ–Ω—à–∏–π —á–∏—Å–ª–æ nn. –ó –ø–µ—Ä—à–æ–≥–æ —Ä—è–¥–∫–∞ —Ñ–æ—Ä–º—É—î—Ç—å—Å—è –ø–µ—Ä–µ–ª—ñ–∫.

–†–µ–∞–ª—ñ–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü—ñ—é chunked(), —è–∫–∞ –ø—Ä–∏–π–º–∞—î –Ω–∞ –≤—Ö—ñ–¥ —Å–ø–∏—Å–æ–∫ —Ç–∞ —á–∏—Å–ª–æ, —â–æ –∑–∞–¥–∞—î —Ä–æ–∑–º—ñ—Ä —á–∞–Ω–∫–∞ (—à–º–∞—Ç–∫–∞), –∞ –ø–æ–≤–µ—Ä—Ç–∞—î —Å–ø–∏—Å–æ–∫ —ñ–∑ —á–∞–Ω–∫—ñ–≤ –≤–∫–∞–∑–∞–Ω–æ—ó –¥–æ–≤–∂–∏–Ω–∏.

<b>–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ù–∞ –≤—Ö—ñ–¥ –ø—Ä–æ–≥—Ä–∞–º—ñ –ø–æ–¥–∞—î—Ç—å—Å—è —Ä—è–¥–æ–∫ —Ç–µ–∫—Å—Ç—É, —â–æ –º—ñ—Å—Ç–∏—Ç—å —Å–∏–º–≤–æ–ª–∏, –≤—ñ–¥–æ–∫—Ä–µ–º–ª–µ–Ω—ñ —Å–∏–º–≤–æ–ª–æ–º –ø—Ä–æ–ø—É—Å–∫—É —Ç–∞ —á–∏—Å–ª–æ nn –Ω–∞ –æ–∫—Ä–µ–º–æ–º—É —Ä—è–¥–∫—É.

<b>–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ü—Ä–æ–≥—Ä–∞–º–∞ –º–∞—î –≤–∏–≤–µ—Å—Ç–∏ –≤–∫–∞–∑–∞–Ω–∏–π —Å–ø–∏—Å–æ–∫.

<b>–ü—Ä–∏–º—ñ—Ç–∫–∞</b>. –ù–µ –∑–∞–±—É–¥—å—Ç–µ –≤–∏–∫–ª–∏–∫–∞—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—é chunked(), —â–æ–± –≤–∏–≤–µ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç üòÄ.

```python
# put your python code here
text = input().split(' ')
n = int(input())
lst = []
temp = []
 
for k, item in enumerate(text):
    temp.append(item)
    if len(temp) == n:
        lst.append(temp)
        temp = []
if (len(temp) % n) != 0:
    lst.append(temp)
print(lst)
```

<h3 align='center'>–ü—ñ–¥–ø–∏—Å–∫–∏ —Å–ø–∏—Å–∫—É üå∂Ô∏èüå∂Ô∏è</h3>
–ü—ñ–¥—Å–ø–∏—Å–æ–∫ ‚Äî —Ü–µ —á–∞—Å—Ç–∏–Ω–∞ —ñ–Ω—à–æ–≥–æ —Å–ø–∏—Å–∫—É. –ü—ñ–¥—Å–ø–∏—Å–æ–∫ –º–æ–∂–µ –º—ñ—Å—Ç–∏—Ç–∏ –æ–¥–∏–Ω –µ–ª–µ–º–µ–Ω—Ç, –∫—ñ–ª—å–∫–∞ —ñ –Ω–∞–≤—ñ—Ç—å –∂–æ–¥–Ω–æ–≥–æ. –ù–∞–ø—Ä–∏–∫–ª–∞–¥, [1], [2], [3] —Ç–∞ [4] - –ø—ñ–¥–ø–∏—Å–∫–∏ —Å–ø–∏—Å–∫—É [1, 2, 3, 4]. –°–ø–∏—Å–æ–∫ [2, 3] ‚Äî –ø—ñ–¥—Å–ø–∏—Å–æ–∫ —Å–ø–∏—Å–∫—É [1, 2, 3, 4], –∞–ª–µ —Å–ø–∏—Å–æ–∫ [2, 4] –Ω–µ –ø—ñ–¥—Å–ø–∏—Å–æ–∫ —Å–ø–∏—Å–∫—É [1, 2, 3, 4], –æ—Å–∫—ñ–ª—å–∫–∏ –µ–ª–µ–º–µ–Ω—Ç–∏ 22 —Ç–∞ 44 —É –¥—Ä—É–≥–æ–º—É —Å–ø–∏—Å–∫—É –Ω–µ —Å—É–º—ñ–∂–Ω—ñ . –ü–æ—Ä–æ–∂–Ω—ñ–π —Å–ø–∏—Å–æ–∫ ‚Äì –ø—ñ–¥—Å–ø–∏—Å–æ–∫ –±—É–¥—å-—è–∫–æ–≥–æ —Å–ø–∏—Å–∫—É. –°–∞–º —Å–ø–∏—Å–æ–∫ - –ø—ñ–¥—Å–ø–∏—Å–æ–∫ —Å–µ–±–µ, —Ç–æ–±—Ç–æ —Å–ø–∏—Å–æ–∫ [1, 2, 3, 4] –ø—ñ–¥—Å–ø–∏—Å–æ–∫ —Å–ø–∏—Å–∫—É [1, 2, 3, 4].

–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ø–æ–¥–∞—î—Ç—å—Å—è —Ä—è–¥–æ–∫ —Ç–µ–∫—Å—Ç—É, —â–æ –º—ñ—Å—Ç–∏—Ç—å —Å–∏–º–≤–æ–ª–∏. –ó —Ü—å–æ–≥–æ —Ä—è–¥–∫–∞ —Ñ–æ—Ä–º—É—î—Ç—å—Å—è —Å–ø–∏—Å–æ–∫. –ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ –≤–∏–≤–æ–¥–∏—Ç—å —Å–ø–∏—Å–æ–∫, —â–æ –º—ñ—Å—Ç–∏—Ç—å —É—Å—ñ –º–æ–∂–ª–∏–≤—ñ –ø—ñ–¥–ø–∏—Å–∫–∏ —Å–ø–∏—Å–∫—É, –≤–∫–ª—é—á–∞—é—á–∏ –ø—É—Å—Ç–∏–π —Å–ø–∏—Å–æ–∫.

<b>–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ø–æ–¥–∞—î—Ç—å—Å—è —Ä—è–¥–æ–∫ —Ç–µ–∫—Å—Ç—É, —â–æ –º—ñ—Å—Ç–∏—Ç—å —Å–∏–º–≤–æ–ª–∏, –≤—ñ–¥–æ–∫—Ä–µ–º–ª–µ–Ω—ñ —Å–∏–º–≤–æ–ª–æ–º –ø—Ä–æ–ø—É—Å–∫—É.

<b>–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ü—Ä–æ–≥—Ä–∞–º–∞ –ø–æ–≤–∏–Ω–Ω–∞ –≤–∏–≤–µ—Å—Ç–∏ –≤–∫–∞–∑–∞–Ω–∏–π —Å–ø–∏—Å–æ–∫, —â–æ –º—ñ—Å—Ç–∏—Ç—å —É—Å—ñ –º–æ–∂–ª–∏–≤—ñ –ø—ñ–¥–ø–∏—Å–∫–∏, –≤–∫–ª—é—á–∞—é—á–∏ –ø–æ—Ä–æ–∂–Ω—ñ–π —Å–ø–∏—Å–æ–∫ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ –ø—Ä–∏–∫–ª–∞–¥—ñ–≤.

<b>–ü—Ä–∏–º—ñ—Ç–∫–∞</b>. –ü–æ—Ä—è–¥–æ–∫ —Å–ø–∏—Å–∫—ñ–≤ –æ–¥–Ω–∞–∫–æ–≤–æ—ó –¥–æ–≤–∂–∏–Ω–∏ –º–∞—î –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—Ç–∏ –ø–æ—Ä—è–¥–∫—É —ó—Ö–Ω—å–æ–≥–æ –≤—Ö–æ–¥–∂–µ–Ω–Ω—è –¥–æ –æ—Å–Ω–æ–≤–Ω–æ–≥–æ —Å–ø–∏—Å–∫—É.

```python
# put your python code here
text = input().split(' ')
lst = []
result = [[]]
for i in range(len(text)):
    for j in range(len(text)):
        lst = text[j:i + j + 1]
        if len(lst) == i + 1:
            result.append(lst)
print(result)
```
