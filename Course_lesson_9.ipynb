<h3 align='center'>Вершина параболи</h3>
Рівняння параболи має вигляд y = ax<sup>2</sup> + bx + c Напишіть програму, яка за введеними значеннями a, b, c визначає та виводить вершину параболи.

<b>Формат вхідних даних</b><br>
На вхід до програми подаються три цілих числа, кожне на окремому рядку.

<b>Формат вихідних даних</b><br>
Програма має вивести координати вершини параболи.

```python
a = int(input())
b = int(input())
c = int(input())
x = (-b/(2*a))
y = ((4*a*c-b**2)/(4*a))
print('')
print((x, y))
```

<h3 align='center'>Конкурсний відбір</h3>
Напишіть програму, яка виводить список хорошистів та відмінників у класі.

<b>Формат вхідних даних</b><br>
На вхід програмі подається натуральне число nn, далі слідує nn рядків з прізвищем школяра та його оцінкою на кожній з них.

<b>Формат вихідних даних</b><br>
Програма має вивести спочатку всі введені рядки з прізвищами та оцінками учнів у тому самому порядку. Потім слідує порожній рядок, а потім виводяться рядки з прізвищами та оцінками хорошистів та відмінників (у тому ж порядку).

<b>Примітка 1</b>. Оцінка учня – натуральне число від 1 до 5.

<b>Примітка 2</b>. Гарантується, що в класі є хоча б один хороший – володар оцінки 4, або відмінник – 5.

```python
n = int(input())
lst = [(input().split()) for _ in range(n)]
print('-----------')
for i in lst:
    print(*i)
print('')
for i in lst:
    if int(i[1]) >= 4: print(*i)
```

<h3 align='center'>Послідовність Трибоначчі</h3>
Напишіть програму, яка зчитує натуральне число nn та виводить перші nn чисел послідовності Трибоначчі.

<b>Формат вхідних даних</b><br>
На вхід до програми подається одне число n (n≤100) – кількість членів послідовності.

<b>Формат вихідних даних</b><br>
Програма має вивести члени послідовності Трибоначчі, відокремлені символом пропуску.

<b>Примітка</b>. Послідовність Трибоначчі – послідовність натуральних чисел, де кожне наступне число є сумою трьох попередніх:

```python
# put your python code here
lst = (1, 1, 1)
n = int(input())
for i in range(n):
    a, b, c = lst[i:i+3]
    lst += (a+b+c, )
print(*lst[:n])
```
