<h3 align='center'>Кожен n-ий елемент</h3>
На вхід до програми подається рядок тексту, що містить символи та число n. З цього рядка формується список. Напишіть програму, яка поділяє список на вкладені підписки так, що n послідовних елементів належать різним підпискам.

<b>Формат вхідних даних</b><br>
На вхід програмі подається рядок тексту, що містить символи, відокремлені символом пробілу та число n на окремому рядку.

<b>Формат вихідних даних</b><br>
Програма має вивести вказаний список.

<b>Примітка</b>. Графічна ілюстрація для 1 тестів:

```python
text = input().split()
n = int(input())
lst = []
for i in range(n):
    lst.append(text[i::n])
print('')
print(lst)
```

<h3 align='center'>Транспонування матриці</h3>

На вхід до програми подається натуральне число nn — кількість рядків і стовпців у матриці, потім елементи матриці.

<b>Примітка 2</b>. Ви можете вирішити завдання без використання допоміжного списку.

<h3 align='center'>Сніжинка</h3>
На вхід до програми подається непарне натуральне число nn. Напишіть програму, яка створює матрицю розміром n×n заповнивши її символами . Потім заповніть символами * середній рядок та стовпець матриці, головну та побічну діагональ матриці. Виведіть отриману матрицю на екран, розділяючи елементи пробілами.

<b>Формат вхідних даних</b><br>
На вхід до програми подається непарне натуральне число n, (n≥3) — кількість рядків і стовпців у матриці.

<b>Формат вихідних даних</b><br>
Програма повинна вивести матрицю відповідно до умови завдання.

```python
# put your python code here
n = int(input())
matrix = [['.']*n for i in range(n)]
for i in range(len(matrix)):
    matrix[i][i] = '*'
    matrix[i][-i-1] = '*'
    matrix[i][n//2] = '*'
    matrix[n//2][i] = '*'
for i in range(len(matrix)):
    print(*matrix[i])
```

<h3 align='center'>Симетрична матриця</h3>
Напишіть програму перевірки симетричності квадратної матриці щодо побічної діагоналі.

<b>Формат вхідних даних</b><br>
На вхід до програми подається натуральне число n — кількість рядків і стовпців у матриці, потім елементи матриці.

<b>Формат вихідних даних</b><br>
Програма повинна вивести YES, якщо матриця симетрична, і слово NO інакше.

```python
# put your python code here
n = int(input())
matrix = [list(map(int, input().split())) for i in range(n)]
flag = True
for i in range(n):
    for j in range(n):
        if matrix[i][j] == matrix[n - j - 1 ][n - i - 1]: continue
        else: flag = False
print('')
if flag:print('YES')
else: print('NO')
```

<h3 align='center'>Латинский квадрат 🌶️</h3>
Латинским квадратом порядка n называется квадратная матрица размером n×n, каждая строка и каждый столбец которой содержат все числа от 1 до n. Напишите программу, которая проверяет, является ли заданная квадратная матрица латинским квадратом.

<b>Формат входных данных</b><br>
На вход программе подаётся натуральное число n — количество строк и столбцов в матрице, затем элементы матрицы: n строк, по nn чисел в каждой, разделённые пробелами.

<b>Формат выходных данных</b><br>
Программа должна вывести слово YES, если матрица является латинским квадратом, и слово NO, если не является.

```python
n = int(input())
matrix = [[int(i) for i in input().split()] for _ in range(n)]
numbers = list(range(1, n + 1))
result = 'YES'

for i in range(n):
    row_nums = sorted(matrix[i])
    col_nums = sorted([matrix[j][i] for j in range(n)])
    if row_nums != numbers or col_nums != numbers:
        result = 'NO'
        break
print('')       
print(result)
```

<h3 align='center'>Ходи ферзя</h3>
На шахівниці 8×8 стоїть ферзь. Позначте положення ферзя на дошці та всі клітини, які б'є ферзь. Клітку, де стоїть ферзь, позначте буквою Q, клітини, які б'є ферзь, позначте символами *, решту клітин заповніть крапками.

<b>Формат вхідних даних</b><br>
На вхід програми подаються координати ферзя на шахівниці у шаховій нотації (тобто у вигляді e4, де спочатку записується номер стовпця (літера від a до h, зліва направо), потім номер рядка (цифра від 1 до 8, знизу догори)).

<b>Формат вихідних даних</b><br>
Програма має вивести на екран зображення дошки, розділяючи елементи пробілами.

```python
# put your python code here
xy = input()
x1 = '87654321'.index(xy[1])
y1 = 'abcdefgh'.index(xy[0])
matrix = [['.']*8 for _ in range(8)]
for x2 in range(8):
    for y2 in range(8):
        if (abs(x1 - x2) == abs(y1 - y2)) or (x1 == x2 or y1 == y2):
            matrix[x2][y2] = '*'
matrix[x1][y1] = 'Q'
for i in range(8):
    print(*matrix[i])
```

<h3 align='center'>Діагоналі паралельні головній</h3>
На вхід до програми подається натуральне число n. Напишіть програму, яка створює матрицю розміром n×n та заповнює її за наступним правилом:<br><br>

<li>на головній діагоналі дома кожного елемента має стояти число 0;
<li>на двох діагоналях, що прилягають до головної, число 1;
<li>на наступних двох діагоналях число 2 і т.д.
<br><br>
<b>Формат вхідних даних</b><br>
На вхід до програми подається натуральне число n — кількість рядків і стовпців у матриці.

<b>Формат вихідних даних</b><br>
Програма повинна вивести матрицю відповідно до умови завдання.

```python
# put your python code here
n = int(input())
matrix = [[0]*n for _ in range(n)]
for i in range(n):
    for j in range(n):
        if i != j:
            matrix[i][j] = abs(i - j)
print('')        
for i in range(n):
    print(*matrix[i])
```
