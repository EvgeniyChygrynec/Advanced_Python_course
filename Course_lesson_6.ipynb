<h3 align='center'>–®–∞—Ö—ñ–≤–Ω–∏—Ü—è</h3>
–ù–∞ –≤—Ö—ñ–¥ –ø—Ä–æ–≥—Ä–∞–º—ñ –ø–æ–¥–∞—é—Ç—å—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ñ —á–∏—Å–ª–∞ n —ñ m. –ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è –º–∞—Ç—Ä–∏—Ü—ñ —Ä–æ–∑–º—ñ—Ä–æ–º n \times mn√óm, –∑–∞–ø–æ–≤–Ω–∏–≤—à–∏ —ó—ó —Å–∏–º–≤–æ–ª–∞–º–∏ . —ñ * —É —à–∞—Ö–æ–≤–æ–º—É –ø–æ—Ä—è–¥–∫—É. –£ –ª—ñ–≤–æ–º—É –≤–µ—Ä—Ö–Ω—å–æ–º—É –∫—É—Ç–∫—É –º–∞—î —Å—Ç–æ—è—Ç–∏ –∫—Ä–∞–ø–∫–∞. –í–∏–≤–µ–¥—ñ—Ç—å –æ—Ç—Ä–∏–º–∞–Ω—É –º–∞—Ç—Ä–∏—Ü—é –Ω–∞ –µ–∫—Ä–∞–Ω, —Ä–æ–∑–¥—ñ–ª—è—é—á–∏ –µ–ª–µ–º–µ–Ω—Ç–∏ –ø—Ä–æ–±—ñ–ª–∞–º–∏.

<b>–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ù–∞ –≤—Ö—ñ–¥ –ø—Ä–æ–≥—Ä–∞–º—ñ –Ω–∞ –æ–¥–Ω–æ–º—É —Ä—è–¥–∫—É –ø–æ–¥–∞—é—Ç—å—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ñ —á–∏—Å–ª–∞ nn —ñ mm - –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ä—è–¥–∫—ñ–≤ —ñ —Å—Ç–æ–≤–ø—Ü—ñ–≤ —É –º–∞—Ç—Ä–∏—Ü—ñ.

<b>–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ü—Ä–æ–≥—Ä–∞–º–∞ –º–∞—î –≤–∏–≤–µ—Å—Ç–∏ –º–∞—Ç—Ä–∏—Ü—é, –æ–ø–∏—Å–∞–Ω—É –∑–∞ —É–º–æ–≤–∏ –∑–∞–≤–¥–∞–Ω–Ω—è.

```python
# put your python code here
n = list(map(int, input().split()))
matrix = [['.']*n[1] for _ in range(n[0])]
for i in range(n[0]):
    for j in range(n[1]):
        if (i % 2 == 0) and (j % 2 == 1):
            matrix[i][j] = '*'
        elif (i % 2 == 1) and (j % 2 == 0):
            matrix[i][j] = '*'
for i in range(n[0]):
    print(*matrix[i])
```

<h3 align='center'>–ü–æ–±–æ—á–Ω–∞—è –¥–∏–∞–≥–æ–Ω–∞–ª—å</h3>
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ nn. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å–æ–∑–¥–∞–µ—Ç –º–∞—Ç—Ä–∏—Ü—É —Ä–∞–∑–º–µ—Ä–æ–º n√ón –∏ –∑–∞–ø–æ–ª–Ω—è–µ—Ç –µ—ë –ø–æ —Å–ª–µ–¥—É—é—â–µ–º—É –ø—Ä–∞–≤–∏–ª—É:<br><br>

<li>—á–∏—Å–ª–∞ –Ω–∞ –ø–æ–±–æ—á–Ω–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏ —Ä–∞–≤–Ω—ã 1;
<li>—á–∏—Å–ª–∞, —Å—Ç–æ—è—â–∏–µ –≤—ã—à–µ —ç—Ç–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏, —Ä–∞–≤–Ω—ã 0;
<li>—á–∏—Å–ª–∞, —Å—Ç–æ—è—â–∏–µ –Ω–∏–∂–µ —ç—Ç–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏, —Ä–∞–≤–Ω—ã 2.<br>

–ü–æ–ª—É—á–µ–Ω–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É –≤—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ —ç–∫—Ä–∞–Ω. –ß–∏—Å–ª–∞ –≤ —Å—Ç—Ä–æ–∫–µ —Ä–∞–∑–¥–µ–ª—è–π—Ç–µ –æ–¥–Ω–∏–º –ø—Ä–æ–±–µ–ª–æ–º.

<b>–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö</b><br>
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –∏ —Å—Ç–æ–ª–±—Ü–æ–≤ –≤ –º–∞—Ç—Ä–∏—Ü–µ.

<b>–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö</b><br>
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –º–∞—Ç—Ä–∏—Ü—É –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

<b>–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ</b>. –ü–æ–±–æ—á–Ω–∞—è –¥–∏–∞–≥–æ–Ω–∞–ª—å ‚Äî —ç—Ç–æ –¥–∏–∞–≥–æ–Ω–∞–ª—å, –∏–¥—É—â–∞—è –∏–∑ –ø—Ä–∞–≤–æ–≥–æ –≤–µ—Ä—Ö–Ω–µ–≥–æ –≤ –ª–µ–≤—ã–π –Ω–∏–∂–Ω–∏–π —É–≥–æ–ª –º–∞—Ç—Ä–∏—Ü—ã.

```python
# put your python code here
n = int(input())
matrix = [[0]*n for _ in range(n)]
for i in range(n):
    for j in range(n):
        if (i + j) == (n - 1):
            matrix[i][j] = 1
        elif (i +j) >= n:
            matrix[i][j] = 2
        else:
            matrix[i][j] = 0
for i in range(n):
    print(*matrix[i])
```

<b>–ü—Ä–∏–º—ñ—Ç–∫–∞</b>. –î–ª—è –≤–∏–≤–µ–¥–µ–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –º–∞—Ç—Ä–∏—Ü—ñ —è–∫ —É –ø—Ä–∏–∫–ª–∞–¥–∞—Ö, –≤—ñ–¥–≤–æ–¥—å—Ç–µ —Ä—ñ–≤–Ω–æ 3 —Å–∏–º–≤–æ–ª–∏ –Ω–∞ –∫–æ–∂–µ–Ω –µ–ª–µ–º–µ–Ω—Ç. –î–ª—è —Ü—å–æ–≥–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ —Ä—è–¥–∫–æ–≤–∏–π –º–µ—Ç–æ–¥ ljust(). –ú–æ–∂–Ω–∞ –æ–±—ñ–π—Ç–∏—Å—è —ñ –±–µ–∑ ljust(), —Å–∏—Å—Ç–µ–º–∞ –ø—Ä–∏–π–º–µ —ñ —Ç–∞–∫–µ —Ä—ñ—à–µ–Ω–Ω—è üòá

```python
# put your python code here
n = list(map(int, input().split()))
matrix = [[0]*n[1] for _ in range(n[0])]
count = 1
for i in range(n[0]):
    for j in range(n[1]):
        matrix[i][j] = count
        count += 1
print('')
for i in range(n[0]):
    print(*matrix[i])
```

```python
# put your python code here
n = list(map(int, input().split()))
matrix = [[0]*n[1] for _ in range(n[0])]
count = 1
for i in range(n[1]):
    for j in range(n[0]):
        matrix[j][i] = count
        count += 1
print('')
for i in range(n[0]):
    print(*matrix[i])
```

<h3 align='center'>–ó–∞–ø–æ–≤–Ω–µ–Ω–Ω—è 3</h3>
–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ø–æ–¥–∞—î—Ç—å—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–µ —á–∏—Å–ª–æ n. –ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ —Å—Ç–≤–æ—Ä—é—î –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º n√ón –∑–∞–ø–æ–≤–Ω–∏–≤—à–∏ —ó—ó –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ –∑—Ä–∞–∑–∫–∞.

<b>–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ø–æ–¥–∞—î—Ç—å—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–µ —á–∏—Å–ª–æ n ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ä—è–¥–∫—ñ–≤ —ñ —Å—Ç–æ–≤–ø—Ü—ñ–≤ —É –º–∞—Ç—Ä–∏—Ü—ñ.

<b>–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ü—Ä–æ–≥—Ä–∞–º–∞ –ø–æ–≤–∏–Ω–Ω–∞ –≤–∏–≤–µ—Å—Ç–∏ –∑–∞–∑–Ω–∞—á–µ–Ω—É –º–∞—Ç—Ä–∏—Ü—é –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ –∑—Ä–∞–∑–∫–∞: —Ä–æ–∑–º—ñ—Å—Ç–∏—Ç–∏ –æ–¥–∏–Ω–∏—Ü—ñ –Ω–∞ –≥–æ–ª–æ–≤–Ω—ñ–π —Ç–∞ –ø–æ–±—ñ—á–Ω—ñ–π –¥—ñ–∞–≥–æ–Ω–∞–ª—è—Ö, —Ä–µ—à—Ç—É –ø–æ–∑–∏—Ü—ñ–π –º–∞—Ç—Ä–∏—Ü—ñ –∑–∞–ø–æ–≤–Ω–∏—Ç–∏ –Ω—É–ª—è–º–∏.

<b>–ü—Ä–∏–º—ñ—Ç–∫–∞</b>. –î–ª—è –≤–∏–≤–µ–¥–µ–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –º–∞—Ç—Ä–∏—Ü—ñ —è–∫ —É –ø—Ä–∏–∫–ª–∞–¥–∞—Ö, –≤—ñ–¥–≤–æ–¥—å—Ç–µ —Ä—ñ–≤–Ω–æ 3 —Å–∏–º–≤–æ–ª–∏ –Ω–∞ –∫–æ–∂–µ–Ω –µ–ª–µ–º–µ–Ω—Ç. –î–ª—è —Ü—å–æ–≥–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ —Ä—è–¥–∫–æ–≤–∏–π –º–µ—Ç–æ–¥ ljust(). –ú–æ–∂–Ω–∞ –æ–±—ñ–π—Ç–∏—Å—è —ñ –±–µ–∑ ljust(), —Å–∏—Å—Ç–µ–º–∞ –ø—Ä–∏–π–º–µ —ñ —Ç–∞–∫–µ —Ä—ñ—à–µ–Ω–Ω—è üòá

```python
# put your python code here
n = int(input())
matrix = [[0]*n for _ in range(n)]
for i in range(n):
    for j in range(n):
        if (i == j):
            matrix[i][j] = 1
            matrix[i][n-j-1] = 1
print('')       
for i in range(n):
    print(*matrix[i])
```

<b>–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ø–æ–¥–∞—î—Ç—å—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–µ —á–∏—Å–ª–æ n ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ä—è–¥–∫—ñ–≤ —ñ —Å—Ç–æ–≤–ø—Ü—ñ–≤ —É –º–∞—Ç—Ä–∏—Ü—ñ.

<b>–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ü—Ä–æ–≥—Ä–∞–º–∞ –ø–æ–≤–∏–Ω–Ω–∞ –≤–∏–≤–µ—Å—Ç–∏ –≤–∫–∞–∑–∞–Ω—É –º–∞—Ç—Ä–∏—Ü—é –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ –∑—Ä–∞–∑–∫–∞.

<b>–ü—Ä–∏–º—ñ—Ç–∫–∞</b>. –î–ª—è –≤–∏–≤–µ–¥–µ–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –º–∞—Ç—Ä–∏—Ü—ñ —è–∫ —É –ø—Ä–∏–∫–ª–∞–¥–∞—Ö, –≤—ñ–¥–≤–æ–¥—å—Ç–µ —Ä—ñ–≤–Ω–æ 3 —Å–∏–º–≤–æ–ª–∏ –Ω–∞ –∫–æ–∂–µ–Ω –µ–ª–µ–º–µ–Ω—Ç. –î–ª—è —Ü—å–æ–≥–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ —Ä—è–¥–∫–æ–≤–∏–π –º–µ—Ç–æ–¥ ljust(). –ú–æ–∂–Ω–∞ –æ–±—ñ–π—Ç–∏—Å—è —ñ –±–µ–∑ ljust(), —Å–∏—Å—Ç–µ–º–∞ –ø—Ä–∏–π–º–µ —ñ —Ç–∞–∫–µ —Ä—ñ—à–µ–Ω–Ω—è üòá

```python
# put your python code here
n = int(input())
matrix = [[0]*n for _ in range(n)]
for i in range(n):
    for j in range(n):
        if ((i >= j) and (i >= n - 1 - j)) or ((i <= j) and (i <= n - 1 - j)):
            matrix[i][j] = 1
print('')       
for i in range(n):
    print(*matrix[i])
```

<b>–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ù–∞ –≤—Ö—ñ–¥ –ø—Ä–æ–≥—Ä–∞–º—ñ –Ω–∞ –æ–¥–Ω–æ–º—É —Ä—è–¥–∫—É –ø–æ–¥–∞—é—Ç—å—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ñ —á–∏—Å–ª–∞ n —ñ m - –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ä—è–¥–∫—ñ–≤ —ñ —Å—Ç–æ–≤–ø—Ü—ñ–≤ —É –º–∞—Ç—Ä–∏—Ü—ñ.

<b>–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ü—Ä–æ–≥—Ä–∞–º–∞ –ø–æ–≤–∏–Ω–Ω–∞ –≤–∏–≤–µ—Å—Ç–∏ –≤–∫–∞–∑–∞–Ω—É –º–∞—Ç—Ä–∏—Ü—é –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ –∑—Ä–∞–∑–∫–∞.

<b>–ü—Ä–∏–º—ñ—Ç–∫–∞</b>. –î–ª—è –≤–∏–≤–µ–¥–µ–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –º–∞—Ç—Ä–∏—Ü—ñ —è–∫ —É –ø—Ä–∏–∫–ª–∞–¥–∞—Ö, –≤—ñ–¥–≤–æ–¥—å—Ç–µ —Ä—ñ–≤–Ω–æ 3 —Å–∏–º–≤–æ–ª–∏ –Ω–∞ –∫–æ–∂–µ–Ω –µ–ª–µ–º–µ–Ω—Ç. –î–ª—è —Ü—å–æ–≥–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ —Ä—è–¥–∫–æ–≤–∏–π –º–µ—Ç–æ–¥ ljust(). –ú–æ–∂–Ω–∞ –æ–±—ñ–π—Ç–∏—Å—è —ñ –±–µ–∑ ljust(), —Å–∏—Å—Ç–µ–º–∞ –ø—Ä–∏–π–º–µ —ñ —Ç–∞–∫–µ —Ä—ñ—à–µ–Ω–Ω—è üòá

```python
# put your python code here
n,m = [int(i) for i in input().split()]
matrix = [[0] * m for _ in range(n)]
for i in range(n):
    for j in range(m):
        matrix[i][j] =  (i  + j) % m + 1
for i in range(n):
        print(*matrix[i])
```

<b>–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ù–∞ –≤—Ö—ñ–¥ –ø—Ä–æ–≥—Ä–∞–º—ñ –Ω–∞ –æ–¥–Ω–æ–º—É —Ä—è–¥–∫—É –ø–æ–¥–∞—é—Ç—å—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ñ —á–∏—Å–ª–∞ n —ñ m - –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ä—è–¥–∫—ñ–≤ —ñ —Å—Ç–æ–≤–ø—Ü—ñ–≤ —É –º–∞—Ç—Ä–∏—Ü—ñ.
<b>–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ü—Ä–æ–≥—Ä–∞–º–∞ –ø–æ–≤–∏–Ω–Ω–∞ –≤–∏–≤–µ—Å—Ç–∏ –≤–∫–∞–∑–∞–Ω—É –º–∞—Ç—Ä–∏—Ü—é –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ –∑—Ä–∞–∑–∫–∞.

<b>–ü—Ä–∏–º—ñ—Ç–∫–∞</b>. –î–ª—è –≤–∏–≤–µ–¥–µ–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –º–∞—Ç—Ä–∏—Ü—ñ —è–∫ —É –ø—Ä–∏–∫–ª–∞–¥–∞—Ö, –≤—ñ–¥–≤–æ–¥—å—Ç–µ —Ä—ñ–≤–Ω–æ 3 —Å–∏–º–≤–æ–ª–∏ –Ω–∞ –∫–æ–∂–µ–Ω –µ–ª–µ–º–µ–Ω—Ç. –î–ª—è —Ü—å–æ–≥–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ —Ä—è–¥–∫–æ–≤–∏–π –º–µ—Ç–æ–¥ ljust(). –ú–æ–∂–Ω–∞ –æ–±—ñ–π—Ç–∏—Å—è —ñ –±–µ–∑ ljust(), —Å–∏—Å—Ç–µ–º–∞ –ø—Ä–∏–π–º–µ —ñ —Ç–∞–∫–µ —Ä—ñ—à–µ–Ω–Ω—è üòá

```python
# put your python code here
n = list(map(int, input().split()))
matrix = [[0]*n[1] for _ in range(n[0])]
count = 1
for i in range(n[0]):
    for j in range(n[1]):
        if i % 2 == 0:
            matrix[i][j] = count
            count += 1
        else:
            matrix[i][-j-1] = count
            count += 1
print('')
for i in range(n[0]):
    print(*matrix[i])
```

–ù–∞ –≤—Ö—ñ–¥ –ø—Ä–æ–≥—Ä–∞–º—ñ –ø–æ–¥–∞—é—Ç—å—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ñ —á–∏—Å–ª–∞ n —ñ m. –ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ —Å—Ç–≤–æ—Ä—é—î –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º n√óm –∑–∞–ø–æ–≤–Ω–∏–≤—à–∏ —ó—ó "–¥—ñ–∞–≥–æ–Ω–∞–ª—è–º–∏" –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ –∑—Ä–∞–∑–∫–∞.

<b>–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ù–∞ –≤—Ö—ñ–¥ –ø—Ä–æ–≥—Ä–∞–º—ñ –Ω–∞ –æ–¥–Ω–æ–º—É —Ä—è–¥–∫—É –ø–æ–¥–∞—é—Ç—å—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ñ —á–∏—Å–ª–∞ n —ñ m - –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ä—è–¥–∫—ñ–≤ —ñ —Å—Ç–æ–≤–ø—Ü—ñ–≤ —É –º–∞—Ç—Ä–∏—Ü—ñ.

<b>–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ü—Ä–æ–≥—Ä–∞–º–∞ –ø–æ–≤–∏–Ω–Ω–∞ –≤–∏–≤–µ—Å—Ç–∏ –≤–∫–∞–∑–∞–Ω—É –º–∞—Ç—Ä–∏—Ü—é –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ –∑—Ä–∞–∑–∫–∞.

<b>–ü—Ä–∏–º—ñ—Ç–∫–∞</b>. –î–ª—è –≤–∏–≤–µ–¥–µ–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –º–∞—Ç—Ä–∏—Ü—ñ —è–∫ —É –ø—Ä–∏–∫–ª–∞–¥–∞—Ö, –≤—ñ–¥–≤–æ–¥—å—Ç–µ —Ä—ñ–≤–Ω–æ 3 —Å–∏–º–≤–æ–ª–∏ –Ω–∞ –∫–æ–∂–µ–Ω –µ–ª–µ–º–µ–Ω—Ç. –î–ª—è —Ü—å–æ–≥–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ —Ä—è–¥–∫–æ–≤–∏–π –º–µ—Ç–æ–¥ ljust(). –ú–æ–∂–Ω–∞ –æ–±—ñ–π—Ç–∏—Å—è —ñ –±–µ–∑ ljust(), —Å–∏—Å—Ç–µ–º–∞ –ø—Ä–∏–π–º–µ —ñ —Ç–∞–∫–µ —Ä—ñ—à–µ–Ω–Ω—è üòá

```python
# put your python code here
n = list(map(int, input().split()))
matrix = [[0]*n[1] for _ in range(n[0])]
count = 1
for q in range(n[0]*n[1]):
    for i in range(n[0]):
        for j in range(n[1]):
            if i+j == q:
                matrix[i][j] = count
                count += 1
print('')
for i in range(n[0]):
    print(*matrix[i])
```

<h3 align='center'>–ó–∞–ø–æ–≤–Ω–µ–Ω–Ω—è —Å–ø—ñ—Ä–∞–ª–ª—é üòàüòà</h3>
–ù–∞ –≤—Ö—ñ–¥ –ø—Ä–æ–≥—Ä–∞–º—ñ –ø–æ–¥–∞—é—Ç—å—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ñ —á–∏—Å–ª–∞ n —ñ m. –ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ —Å—Ç–≤–æ—Ä—é—î –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä–æ–º n√óm –∑–∞–ø–æ–≤–Ω–∏–≤—à–∏ —ó—ó "—Å–ø—ñ—Ä–∞–ª–ª—é" –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ –∑—Ä–∞–∑–∫–∞.

<b>–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ù–∞ –≤—Ö—ñ–¥ –ø—Ä–æ–≥—Ä–∞–º—ñ –Ω–∞ –æ–¥–Ω–æ–º—É —Ä—è–¥–∫—É –ø–æ–¥–∞—é—Ç—å—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ñ —á–∏—Å–ª–∞ n —ñ m - –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ä—è–¥–∫—ñ–≤ —ñ —Å—Ç–æ–≤–ø—Ü—ñ–≤ —É –º–∞—Ç—Ä–∏—Ü—ñ.

<b>–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ü—Ä–æ–≥—Ä–∞–º–∞ –º–∞—î –≤–∏–≤–µ—Å—Ç–∏ –º–∞—Ç—Ä–∏—Ü—é –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ –∑—Ä–∞–∑–∫–∞.

<b>–ü—Ä–∏–º—ñ—Ç–∫–∞</b>. –î–ª—è –≤–∏–≤–µ–¥–µ–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –º–∞—Ç—Ä–∏—Ü—ñ —è–∫ —É –ø—Ä–∏–∫–ª–∞–¥–∞—Ö, –≤—ñ–¥–≤–æ–¥—å—Ç–µ —Ä—ñ–≤–Ω–æ 3 —Å–∏–º–≤–æ–ª–∏ –Ω–∞ –∫–æ–∂–µ–Ω –µ–ª–µ–º–µ–Ω—Ç. –î–ª—è —Ü—å–æ–≥–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ —Ä—è–¥–∫–æ–≤–∏–π –º–µ—Ç–æ–¥ ljust(). –ú–æ–∂–Ω–∞ –æ–±—ñ–π—Ç–∏—Å—è —ñ –±–µ–∑ ljust(), —Å–∏—Å—Ç–µ–º–∞ –ø—Ä–∏–π–º–µ —ñ —Ç–∞–∫–µ —Ä—ñ—à–µ–Ω–Ω—è üòá

```python
n, m = map(int, input().split())
i = 1; j = 0; c = 0
a = [[0] * (100) for _ in range(100)]

while c < m * n:
    while a[i][j+1] == 0 and j < m: a[i][j+1] = c+1; j += 1; c += 1
    while a[i+1][j] == 0 and i < n: a[i+1][j] = c+1; i += 1; c += 1
    while a[i][j-1] == 0 and j > 1: a[i][j-1] = c+1; j -= 1; c += 1
    while a[i-1][j] == 0 and i > 1: a[i-1][j] = c+1; i -= 1; c += 1
print('')
for i in range(1, n+1):
    for j in range(1, m+1):
        print(str(a[i][j]).ljust(3), end=' ')
    print()
```
