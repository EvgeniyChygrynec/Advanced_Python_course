<h3 align='center'>–î–æ–º–∞—à–Ω—î –∑–∞–≤–¥–∞–Ω–Ω—è</h3>
–í—á–∏—Ç–µ–ª—å –ø–µ—Ä–µ–≤—ñ—Ä—è—î –¥–æ–º–∞—à–Ω—î –∑–∞–≤–¥–∞–Ω–Ω—è —É –∫–ª–∞—Å—ñ —ñ –æ—Ç—Ä–∏–º–∞–≤ —Ç–∞–∫—ñ –≤—ñ–¥–ø–æ–≤—ñ–¥—ñ: –∑ n —à–∫–æ–ª—è—Ä—ñ–≤ —É m –¥–æ–º–∞—à–Ω—î –∑–∞–≤–¥–∞–Ω–Ω—è –∑'—ó–≤ —Å–æ–±–∞–∫–∞, —É k –≤—ñ–¥–∫–ª—é—á–∏–ª–∏ —Å–≤—ñ—Ç–ª–æ, –∞ p —É—á–Ω—ñ–≤ —Å–ø—ñ—Ç–∫–∞–ª–∏ –æ–±–∏–¥–≤–∞ –Ω–µ—â–∞—Å—Ç—è. –ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ –≤–∏–∑–Ω–∞—á–∞—î, —Å–∫—ñ–ª—å–∫–∏ –ª—é–¥–µ–π –≤–∏–∫–æ–Ω–∞–ª–æ –¥–æ–º–∞—à–Ω—î –∑–∞–≤–¥–∞–Ω–Ω—è.

<b>–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ø–æ–¥–∞—é—Ç—å—Å—è —á–∏—Å–ª–∞ n, m, k, p –∫–æ–∂–Ω–µ –Ω–∞ –æ–∫—Ä–µ–º–æ–º—É —Ä—è–¥–∫—É.

<b>–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ü—Ä–æ–≥—Ä–∞–º–∞ –º–∞—î –≤–∏–≤–µ—Å—Ç–∏ –∫—ñ–ª—å–∫—ñ—Å—Ç—å —É—á–Ω—ñ–≤, —è–∫—ñ –∑—Ä–æ–±–∏–ª–∏ –¥–æ–º–∞—à–Ω—î –∑–∞–≤–¥–∞–Ω–Ω—è.

```python
# put your python code here
n, m, k, p = [int(input()) for i in range(4)]
print('')
print(n-(m-p+p+k-p))
```

<h3 align='center'>–°—Ö—ñ–¥</h3>
–ù–∞ —Å—É–ø—É—Ç–Ω–∏–∫—É ¬´–°—Ö—ñ–¥¬ª –≤—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–æ –ø—Ä–∏–ª–∞–¥ –¥–ª—è –≤–∏–º—ñ—Ä—é–≤–∞–Ω–Ω—è —Å–æ–Ω—è—á–Ω–æ—ó –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—ñ. –©–æ—Ö–≤–∏–ª–∏–Ω–∏ –≤—ñ–Ω –ø–µ—Ä–µ–¥–∞—î –≤ –æ–±—Å–µ—Ä–≤–∞—Ç–æ—Ä—ñ—é –∫–∞–Ω–∞–ª–æ–º –∑–≤'—è–∑–∫—É –ø–æ–∑–∏—Ç–∏–≤–Ω–µ —Ü—ñ–ª–µ —á–∏—Å–ª–æ ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –µ–Ω–µ—Ä–≥—ñ—ó —Å–æ–Ω—è—á–Ω–æ–≥–æ –≤–∏–ø—Ä–æ–º—ñ–Ω—é–≤–∞–Ω–Ω—è. –î–ª—è –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –∞–Ω–∞–ª—ñ–∑—É —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤ –Ω–µ–º–∞—î –Ω–µ–æ–±—Ö—ñ–¥–Ω–æ—Å—Ç—ñ —Ç—Ä–∏–º–∞—Ç–∏ –¥–∞–Ω—ñ, —â–æ –ø–æ–≤—Ç–æ—Ä—é—é—Ç—å—Å—è. –ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ –≤–∏–≤–æ–¥–∏—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Å–≤—ñ–¥—á–µ–Ω—å —Å—É–ø—É—Ç–Ω–∏–∫–∞, –ø—Ä–∏ –≤–∏–¥–∞–ª–µ–Ω–Ω—ñ —è–∫–∏—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç –±—É–¥–µ –ø—Ä–∞–≤–∏–ª—å–Ω–æ –ø—Ä–æ–∞–Ω–∞–ª—ñ–∑–æ–≤–∞–Ω–æ.

<b>–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ø–æ–¥–∞—î—Ç—å—Å—è –æ–¥–∏–Ω —Ä—è–¥–æ–∫, —â–æ –º—ñ—Å—Ç–∏—Ç—å —á–∏—Å–ª–∞ ‚Äì –ø–æ–∫–∞–∑–∞–Ω–Ω—è —Å—É–ø—É—Ç–Ω–∏–∫–∞ ¬´–°—Ö—ñ–¥¬ª. –ß–∏—Å–ª–∞ –≤–∫–∞–∑—É—é—Ç—å—Å—è —á–µ—Ä–µ–∑ –ø—Ä–æ–ø—É—Å–∫ —ñ –º—ñ—Å—Ç—è—Ç—å –ø—Ä–æ–≤—ñ–¥–Ω–∏—Ö –Ω—É–ª—ñ–≤.

<b>–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ü—Ä–æ–≥—Ä–∞–º–∞ –ø–æ–≤–∏–Ω–Ω–∞ –≤–∏–≤–µ—Å—Ç–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø–æ–∫–∞–∑–∞–Ω—å, –ø—ñ—Å–ª—è –≤–∏–¥–∞–ª–µ–Ω–Ω—è —è–∫–∏—Ö –∞–Ω–∞–ª—ñ–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤ –±—É–¥–µ –∑—Ä–æ–±–ª–µ–Ω–æ –ø—Ä–∞–≤–∏–ª—å–Ω–æ.

```python
# put your python code here
n = input().split(' ')
print('')
print(len(n)-len(set(n)))
```

<h3 align='center'>–ú—ñ—Å—Ç–∞</h3>
–¢–∏–º—É—Ä —Ç–∞ –†—É—Å–ª–∞–Ω –≥—Ä–∞—é—Ç—å —É –≥—Ä—É –º—ñ—Å—Ç–∞. –í–æ–Ω–∏ –¥—É–∂–µ –ª—é–±–ª—è—Ç—å —Ü—é –≥—Ä—É —ñ –∑–Ω–∞—é—Ç—å –±–∞–≥–∞—Ç–æ –º—ñ—Å—Ç, –æ—Å–æ–±–ª–∏–≤–æ –¢–∏–º—É—Ä, –æ–¥–Ω–∞–∫ –¥–æ –∫—ñ–Ω—Ü—è –≥—Ä–∏ —á–µ—Ä–µ–∑ —Å–≤—ñ–π –≤—ñ–∫ –∑–∞–±—É–≤–∞—é—Ç—å, —è–∫—ñ –º—ñ—Å—Ç–∞ –≤–∂–µ –Ω–∞–∑–∏–≤–∞–ª–∏.

–ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ –∑—á–∏—Ç—É—î —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—é –ø—Ä–æ –≥—Ä—É —Ç–∞ –ø–æ–≤—ñ–¥–æ–º–ª—è—î —Ö–ª–æ–ø—Ü—è–º, —â–æ —á–µ—Ä–≥–æ–≤–µ –º—ñ—Å—Ç–æ –Ω–∞–∑–≤–∞–Ω–æ –ø–æ–≤—Ç–æ—Ä–Ω–æ.

<b>–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ù–∞ –≤—Ö—ñ–¥ –ø—Ä–æ–≥—Ä–∞–º—ñ –≤ –ø–µ—Ä—à–æ–º—É —Ä—è–¥–∫—É –ø–æ–¥–∞—î—Ç—å—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–µ —á–∏—Å–ª–æ nn ‚Äì –∫—ñ–ª—å–∫—ñ—Å—Ç—å –Ω–∞–∑–≤–∞–Ω–∏—Ö –º—ñ—Å—Ç, —É –Ω–∞—Å—Ç—É–ø–Ω–∏—Ö nn —Ä—è–¥–∫–∞—Ö –≤–≤–æ–¥—è—Ç—å—Å—è –Ω–∞–∑–≤–∞–Ω—ñ –º—ñ—Å—Ç–∞ —Ç–∞ —â–µ –æ–¥–∏–Ω —Ä—è–¥–æ–∫ –∑ –Ω–æ–≤–∏–º, —â–æ–π–Ω–æ –Ω–∞–∑–≤–∞–Ω–∏–º –º—ñ—Å—Ç–æ–º.
    
<b>–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ü—Ä–æ–≥—Ä–∞–º–∞ –ø–æ–≤–∏–Ω–Ω–∞ –≤–∏–≤–µ—Å—Ç–∏ OK, —è–∫—â–æ —Ü–µ –º—ñ—Å—Ç–æ —â–µ –Ω–µ –∑–≥–∞–¥—É–≤–∞–ª–∏, —Ç–∞ REPEAT, —è–∫—â–æ –º—ñ—Å—Ç–æ –≤–∂–µ –±—É–ª–æ –Ω–∞–∑–≤–∞–Ω–æ.

```python
# put your python code here
n = int(input())
lst = []
for _ in range(n+1):
    lst.append(input())
flag = len(lst)-len(set(lst))
print('')
if flag == 0: print('OK')
else: print('REPEAT')
```

<h3 align='center'>–ö–Ω–∏–≥–∏ –Ω–∞ –ø—Ä–æ—á–∏—Ç–∞–Ω–Ω—è</h3>
–†—É—Å–ª–∞–Ω –æ—Ç—Ä–∏–º–∞–≤ –Ω–∞–ø—Ä–∏–∫—ñ–Ω—Ü—ñ –Ω–∞–≤—á–∞–ª—å–Ω–æ–≥–æ —Ä–æ–∫—É —Å–ø–∏—Å–æ–∫ –ª—ñ—Ç–µ—Ä–∞—Ç—É—Ä–∏ –Ω–∞ –ª—ñ—Ç–æ. –¢–µ–ø–µ—Ä –π–æ–º—É —Ç—Ä–µ–±–∞ –∑'—è—Å—É–≤–∞—Ç–∏, —è–∫—ñ –∫–Ω–∏–≥–∏ –∑ —Ü—å–æ–≥–æ —Å–ø–∏—Å–∫—É –º–∞—î. –£ –†—É—Å–ª–∞–Ω–∞ –Ω–∞ –∫–æ–º–ø'—é—Ç–µ—Ä—ñ –≤ —Ç–µ–∫—Å—Ç–æ–≤–æ–º—É —Ñ–∞–π–ª—ñ –∑–∞–ø–∏—Å–∞–Ω—ñ –≤—Å—ñ –∫–Ω–∏–≥–∏ –∑ –π–æ–≥–æ –¥–æ–º–∞—à–Ω—å–æ—ó –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∏ –≤–∏–ø–∞–¥–∫–æ–≤–æ.

–ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ –≤–∏–∑–Ω–∞—á–∞—î –¥–ª—è –∫–æ–∂–Ω–æ—ó –∫–Ω–∏–≥–∏ –∑—ñ —Å–ø–∏—Å–∫—É –Ω–∞ –ø—Ä–æ—á–∏—Ç–∞–Ω–Ω—è, —î –≤–æ–Ω–∞ —É –†—É—Å–ª–∞–Ω–∞ —á–∏ –Ω—ñ.

<b>–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –≤ –ø–µ—Ä—à–æ–º—É —Ä—è–¥–∫—É –ø–æ–¥–∞—î—Ç—å—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–µ —á–∏—Å–ª–æ mm ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –∫–Ω–∏–≥ —É –¥–æ–º–∞—à–Ω—ñ–π –±—ñ–±–ª—ñ–æ—Ç–µ—Ü—ñ –†—É—Å–ª–∞–Ω–∞. –£ –¥—Ä—É–≥–æ–º—É —Ä—è–¥–∫—É –∑–∞–ø–∏—Å–∞–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–µ —á–∏—Å–ª–æ nn ‚Äì –∫—ñ–ª—å–∫—ñ—Å—Ç—å –∫–Ω–∏–≥ —É —Å–ø–∏—Å–∫—É –Ω–∞ –ª—ñ—Ç–æ. –î–∞–ª—ñ –π–¥—É—Ç—å mm —Ä—è–¥–∫—ñ–≤ —ñ–∑ –Ω–∞–∑–≤–∞–º–∏ –∫–Ω–∏–≥ —ñ–∑ –¥–æ–º–∞—à–Ω—å–æ—ó –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∏ —Ç–∞ nn —Ä—è–¥–∫—ñ–≤ –Ω–∞–∑–≤ –∑—ñ —Å–ø–∏—Å–∫—É –Ω–∞ –ª—ñ—Ç–æ.

<b>–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ü—Ä–æ–≥—Ä–∞–º–∞ –ø–æ–≤–∏–Ω–Ω–∞ –≤–∏–≤–µ—Å—Ç–∏ nn —Ä—è–¥–∫—ñ–≤, —É –∫–æ–∂–Ω–æ–º—É –∑ —è–∫–∏—Ö –Ω–∞–ø–∏—Å–∞–Ω–æ —Å–ª–æ–≤–æ YES, —è–∫—â–æ –∫–Ω–∏–≥–∞ –∑–Ω–∞–π–¥–µ–Ω–∞ –≤ –±—ñ–±–ª—ñ–æ—Ç–µ—Ü—ñ, —ñ NO, —è–∫—â–æ –Ω—ñ.

```python
# put your python code here
n, m = [int(input()) for _ in range(2)]
lst = [input() for _ in range(n)]
for _ in range(m):
    text = input()
    if text in lst: print('YES')
    else: print('NO')
```

<h3 align='center'>–î–∏–≤–Ω–µ –∑–∞—Ö–æ–ø–ª–µ–Ω–Ω—è</h3>
–Ø–∫ –≤—ñ–¥–æ–º–æ, –º–∞—Ç–µ–º–∞—Ç–∏–∫–∏ –¥–∏–≤–Ω—ñ –ª—é–¥–∏. –ù–µ —î –≤–∏–Ω—è—Ç–∫–æ–º —ñ –¢–∏–º—É—Ä ‚Äî –∞–≤—Ç–æ—Ä —Ü—å–æ–≥–æ –∫—É—Ä—Å—É. –©–æ–¥–Ω—è –¢–∏–º—É—Ä –≤–∏—Ä—ñ—à—É—î –¥–≤—ñ —Å–∫–ª–∞–¥–Ω—ñ –º–∞—Ç–µ–º–∞—Ç–∏—á–Ω—ñ –∑–∞–≤–¥–∞–Ω–Ω—è. –í–∏—Ä—ñ—à—É—é—á–∏ –ø–µ—Ä—à–µ –∑–∞–≤–¥–∞–Ω–Ω—è, –≤—ñ–Ω –∑–∞–ø–∏—Å—É—î –Ω–∞ –ø–µ—Ä—à–æ–º—É –ª–∏—Å—Ç–∫—É –≤—Å—ñ —á–∏—Å–ª–∞, —è–∫—ñ –≤ –Ω—å–æ–º—É –∑—É—Å—Ç—Ä—ñ—á–∞—é—Ç—å—Å—è. –î–∞–ª—ñ –≤—ñ–Ω —Ä–æ–±–∏—Ç—å –ø–∞—É–∑—É —Ç–∞ –±–µ—Ä–µ—Ç—å—Å—è –∑–∞ –¥—Ä—É–≥–µ –∑–∞–≤–¥–∞–Ω–Ω—è. –ü–æ—Ç—ñ–º –∑–∞–ø–∏—Å—É—î –Ω–∞ –¥—Ä—É–≥–æ–º—É –ª–∏—Å—Ç–∫—É –≤—Å—ñ —á–∏—Å–ª–∞, —è–∫—ñ –≤ –Ω—ñ–π –∑—É—Å—Ç—Ä—ñ—á–∞—é—Ç—å—Å—è. –ü—ñ—Å–ª—è —Ü—å–æ–≥–æ –≤—ñ–Ω –±–µ—Ä–µ —â–µ –æ–¥–∏–Ω –ª–∏—Å—Ç–æ–∫ —ñ –≤–∏–ø–∏—Å—É—î –Ω–∞ –Ω—å–æ–≥–æ –≤—Å—ñ –∑–±—ñ–≥–∞—é—Ç—å—Å—è —á–∏—Å–ª–∞ –∑ –ø–µ—Ä—à–∏—Ö –¥–≤–æ—Ö –ª–∏—Å—Ç–æ—á–∫—ñ–≤. –Ø–∫—â–æ —Ç–∞–∫—ñ —á–∏—Å–ª–∞ —î, –¥–µ–Ω—å –≤–∏–π—à–æ–≤, —è–∫—â–æ –∑–∞–≥–∞–ª—å–Ω–∏—Ö —á–∏—Å–µ–ª –Ω–µ–º–∞—î, –¢–∏–º—É—Ä –≤–≤–∞–∂–∞—î –¥–µ–Ω—å –Ω–µ–≤–¥–∞–ª–∏–º.

–ù–∞–ø–∏—à—ñ—Ç—å –ø—Ä–æ–≥—Ä–∞–º—É, —è–∫–∞ –∑–Ω–∞—Ö–æ–¥–∏—Ç—å –∑–∞–≥–∞–ª—å–Ω—ñ —á–∏—Å–ª–∞ –¥–≤–æ—Ö –ª–∏—Å—Ç–æ—á–∫—ñ–≤ –∞–±–æ –ø–æ–≤—ñ–¥–æ–º–ª—è—î, —â–æ –¥–µ–Ω—å –Ω–µ –≤–¥–∞–≤—Å—è üòè

<b>–§–æ—Ä–º–∞—Ç –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ù–∞ –≤—Ö—ñ–¥ –¥–æ –ø—Ä–æ–≥—Ä–∞–º–∏ –ø–æ–¥–∞—é—Ç—å—Å—è –¥–≤–∞ —Ä—è–¥–∫–∏ –∑ —á–∏—Å–ª–∞–º–∏: —É –ø–µ—Ä—à–æ–º—É —Ä—è–¥–∫—É —á–∏—Å–ª–∞ –∑ –ø–µ—Ä—à–æ–≥–æ –ª–∏—Å—Ç–æ—á–∫–∞, —É –¥—Ä—É–≥–æ–º—É –∑ –¥—Ä—É–≥–æ–≥–æ.

<b>–§–æ—Ä–º–∞—Ç –≤–∏—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö</b><br>
–ü—Ä–æ–≥—Ä–∞–º–∞ –ø–æ–≤–∏–Ω–Ω–∞ –≤–∏–≤–µ—Å—Ç–∏ —á–∏—Å–ª–∞, —â–æ –∑—É—Å—Ç—Ä—ñ–ª–∏—Å—è –Ω–∞ –æ–±–æ—Ö –ª–∏—Å—Ç–∫–∞—Ö —É –≤—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–æ–º—É –ø–æ—Ä—è–¥–∫—É –∑–∞ —Å–ø–∞–¥–∞–Ω–Ω—è–º, –∞–±–æ —Å–ª–æ–≤–æ—Å–ø–æ–ª—É—á–µ–Ω–Ω—è BAD DAY, —è–∫—â–æ —Ç–∞–∫–∏—Ö —á–∏—Å–µ–ª –Ω–µ–º–∞—î.

```python
# put your python code here
n = set(input().split(' '))
m = set(input().split(' '))
print('')
if len(n & m) == 0: print('BAD DAY')
else: print(*sorted(n & m, reverse=True, key=int))
```
